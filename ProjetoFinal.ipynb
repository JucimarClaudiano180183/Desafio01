# 3. Média de Avaliação das Lojas

```python
#Criando a função para encontrar a média
def media_avaliacao(lista: list[0])->float:
   media = round(sum(lista)/len(lista),2)
   return media

media_avaliacao02 = media_avaliacao(loja2["Avaliação da compra"])
media_avaliacao03 = media_avaliacao(loja3["Avaliação da compra"])
media_avaliacao04 = media_avaliacao(loja4["Avaliação da compra"])

print(f'loja02: {media_avaliacao02}')
print(f'loja03: {media_avaliacao03}')
print(f'loja04: {media_avaliacao04}')
```

```python
plt.plot(['loja02'],[media_avaliacao02],'o')
plt.plot(['loja03'],[media_avaliacao03],'o')
plt.plot(['loja04'],[media_avaliacao04],'o')
plt.title('Média de Avaliação das Lojas')
```
